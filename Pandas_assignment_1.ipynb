```python
# 5.Which was the most ordered item? and How many items were ordered?
getitemname=df.groupby('item_name')
df1=getitemname.sum()
df2=df1.sort_values('quantity',ascending=False)
df2.head(1)

    
```

```python
# 6.What was the most ordered item in the choice_description column?
gbchoice=df.groupby('choice_description')
df3=gbchoice.sum()
df4=df3.sort_values('quantity',ascending=False)
df4.head(1)
```

```python
 # 7. Turn the item price into a float
def getprice(strprice):
    strtemp=float(strprice[1:])
    return strtemp

df['item_price']=df['item_price'].apply(getprice)
df.head()
```

```python
# 8. print a data frame with only two columns item_name and item_price
df5=df[['item_name','item_price']]
df5
```

```python
# 9. Delete the duplicates in item_name and quantity
df11=df[['item_name','quantity']]
pd.DataFrame.drop_duplicates(df11)
```

```python
 # 10.select only the products with quantity equals to 1
df[df.quantity==1]
```

```python
# 11.select only the item_name and item_price columns
df6=df[['item_name','item_price']]
df6
```

```python
# 12.sort item from the most to less expensive
getitemprice=df6.groupby('item_price')
df7=getitemprice.sum()
df8=df7.sort_values('item_price',ascending=False)
df8
```
